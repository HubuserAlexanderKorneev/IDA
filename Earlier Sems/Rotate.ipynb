```python
def get_patch(image,crop_shape=((0, 0), (0, 0))):
    return image[crop_shape[0][0]:crop_shape[0][1],crop_shape[1][0]:crop_shape[1][1],:]
```

```python
plt.imshow(get_patch(buf,((0,200),(0,200))))
```
